```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\bharath\OneDrive\Desktop\IRIS.csv'
df = pd.read_csv(file_path)

category_counts = df['OSM_ID'].value_counts()

plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('OSM_ID')
plt.show()
```
