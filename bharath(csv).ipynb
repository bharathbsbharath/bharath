```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\bharath\OneDrive\Desktop\GAMES CSV.csv'
df = pd.read_csv(file_path)

# Make sure the column name matches exactly with the case in your CSV file
category_counts = df['BGMI'].value_counts()

# Create a bar chart
category_counts.plot(kind='bar', rot=45)  # Rotating the x-axis labels for better readability

plt.xlabel('BGMI')
plt.ylabel('Count')
plt.title('BGMI')
plt.tight_layout()  # Ensures the labels fit within the figure area
plt.show()
```
